### **Just Testing combine BIRCH + KMeans**

```python
# Apply BIRCH to compress the data first
birch_model = Birch(threshold=1.5, n_clusters=None)  # Disable final clustering
birch_model.fit(data_combined)

# Get subcluster centers as the compressed data representation
compressed_data = birch_model.subcluster_centers_

# Perform K-Means on compressed data (to determine final clusters)
range_n_clusters = range(2, 11)
inertia_scores = []
silhouette_scores = []

for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init='auto')
    cluster_labels = kmeans.fit_predict(compressed_data)
    inertia_scores.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(compressed_data, cluster_labels))

# Plot results
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(range_n_clusters, inertia_scores, marker='o')
plt.title('Elbow Method (Inertia) - BIRCH + K-Means')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')

plt.subplot(1, 2, 2)
plt.plot(range_n_clusters, silhouette_scores, marker='o', color='orange')
plt.title('Silhouette Scores - BIRCH + K-Means')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')

plt.tight_layout()
plt.show()
```


### **Plot Inertia**
* Inertia adalah total jarak data ke pusat cluster-nya (semakin kecil semakin baik).

* Terlihat saat jumlah cluster bertambah, inertia menurun tajam dari 2 ke 3, tapi setelah itu penurunannya mulai melandai.

* Titik "siku" atau elbow muncul pada Cluster = 3, artinya setelah itu menambah cluster tidak lagi signifikan memperbaiki hasil.

Berdasarkan Elbow Method, jumlah cluster optimal yang disarankan adalah 3, karena penurunan inertia yang signifikan hanya terjadi hingga 3 cluster, setelah itu penurunannya melambat dan tidak efisien lagi.




### **Plot Silhoutte**

* Silhouette Score menunjukkan seberapa baik pemisahan antar cluster (semakin mendekati 1, semakin baik).

* Nilai tertinggi ada saat Cluster = 2 (sekitar 0.53), tapi itu berarti hanya dua kelompok besar yang terbentuk (kurang granular untuk segmentasi).

* Setelahnya, skor menurun tajam saat cluster bertambah, yang artinya cluster mulai tumpang tindih dan tidak terpisah dengan baik.

Dilihat dari Silhouette Score, pembentukan cluster paling baik terjadi saat jumlah cluster adalah 2, dengan skor sekitar 0.53. Namun, untuk keperluan segmentasi yang lebih granular, jumlah cluster 3 dapat menjadi kompromi meskipun skor Silhouette sedikit menurun.

Jika fokus pada kualitas cluster yang sangat terpisah → gunakan 2 cluster (skor Silhouette tertinggi).

Jika ingin segmentasi yang lebih praktis dan bermanfaat untuk bisnis → gunakan 3 cluster, karen masih dalam range skor Silhouette yang layak,didukung oleh Elbow Method,dapat memunculkan lebih banyak insight untuk targeting produk

```python
from sklearn.decomposition import PCA

# Get best number of clusters based on silhouette score
best_n_clusters = range_n_clusters[np.argmax(silhouette_scores)]

# Final clustering using K-Means with best number of clusters
final_kmeans = KMeans(n_clusters=best_n_clusters, random_state=42, n_init='auto')
final_labels = final_kmeans.fit_predict(compressed_data)

#Reduce dimensions for visualization using PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(compressed_data)

#Visualize the clustering result
plt.figure(figsize=(8, 6))
scatter = plt.scatter(pca_result[:, 0], pca_result[:, 1], c=final_labels, cmap='Set1', s=50)
plt.title(f'Cluster Visualization (BIRCH + K-Means, k={best_n_clusters})')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.colorbar(scatter, label='Cluster')
plt.grid(True)
plt.show()

best_n_clusters, np.max(silhouette_scores)
```
